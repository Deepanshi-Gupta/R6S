```python
# For notebook environments without OpenCV GUI support (cv2.imshow), display frames inline using matplotlib.
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import cv2
import easyocr
from PIL import Image

# Video path and ROI coordinates
video_path = r"C:\Users\aayus\Downloads\OCR_testing.mp4"
LEFT = 950  # fixed left coordinate as requested
PROCESS_EVERY_N_FRAMES = 10  # process every nth frame

# Initialize EasyOCR reader once (avoid reloading model if already available)
try:
    reader
except NameError:
    reader = easyocr.Reader(['en'])

# Open video
cap = cv2.VideoCapture(video_path)
frame_count = 0
fps = cap.get(cv2.CAP_PROP_FPS) or 25  # fallback if fps is 0
# We'll collect all detected text entries and print a final summary to the terminal after the loop
collected_texts = []  # list of (frame, timestamp_sec, text, conf)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Process only every Nth frame
    if frame_count % PROCESS_EVERY_N_FRAMES != 0:
        frame_count += 1
        continue

    # Calculate timestamp (in seconds)
    timestamp_sec = frame_count / fps if fps > 0 else 0.0

    # Compute ROI according to your specification:
    # - left = 950
    # - right = right-most end of the frame (use frame width)
    # - top/bottom = from the vertical middle of the frame take +/- 250 pixels (clamped to frame bounds)
    h, w = frame.shape[:2]
    left = LEFT
    right = w  # right-most end of frame (slice uses exclusive end)
    mid_y = h // 2
    top = max(0, mid_y - 250)
    bottom = min(h, mid_y + 250)

    # Validate ROI
    if left >= right or top >= bottom:
        print(f"Invalid ROI for frame {frame_count}: left={left}, right={right}, top={top}, bottom={bottom}. Stopping.")
        break

    # Crop ROI from current frame using dynamic ROI
    roi_img = frame[top:bottom, left:right]

    # OCR on ROI (convert BGR to RGB for EasyOCR)
    if roi_img.size == 0:
        frame_count += 1
        continue
    roi_rgb = cv2.cvtColor(roi_img, cv2.COLOR_BGR2RGB)
    results = reader.readtext(roi_rgb, detail=1)

    # Overlay detected text on the ROI for visualization and collect results for terminal summary
    for bbox, text, conf in results:
        print(f"Frame {frame_count} | Time: {timestamp_sec:.2f}s | Text: '{text}' | Confidence: {conf:.3f}")
        collected_texts.append((frame_count, timestamp_sec, text, conf))
        # bbox points are given relative to the roi image passed to EasyOCR
        try:
            pt1 = tuple(map(int, bbox[0]))
            pt2 = tuple(map(int, bbox[2]))
            cv2.rectangle(roi_img, pt1, pt2, (0,255,0), 2)
            cv2.putText(roi_img, text, pt1, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,0), 2)
        except Exception:
            pass

    # Display inline in the notebook (works without OpenCV GUI)
    clear_output(wait=True)
    plt.figure(figsize=(8, 4))
    plt.imshow(cv2.cvtColor(roi_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    display(plt.gcf())
    plt.close()

    frame_count += 1

cap.release()

# After processing, print a full summary of all detected texts to the terminal
print("\n=== OCR SUMMARY ===")
if collected_texts:
    print(f"Total detections: {len(collected_texts)}")
    for frm, tsec, txt, conf in collected_texts:
        print(f"Frame {frm} | Time {tsec:.2f}s | Text: '{txt}' | Conf: {conf:.3f}")
else:
    print("No text detected in the processed frames.")
```

```python
import cv2
from ultralytics import YOLO
import datetime

video_path = r"C:\Users\aayus\Downloads\R6S_2.mp4"
model_path = r"C:\Users\aayus\Downloads\Yolo11_cross.pt"
model_path2 = r"C:\Users\aayus\Downloads\yolo11_bomb.pt"
skip_frames = 4  # Process every 5th frame

# Initialize YOLO models
model1 = YOLO(model_path)
model2 = YOLO(model_path2)

cap = cv2.VideoCapture(video_path)
frame_count = 0
fps = cap.get(cv2.CAP_PROP_FPS)

print("-" * 50)
print(f"Analysis Started at: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("-" * 50)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count % (skip_frames + 1) != 0:
        frame_count += 1
        continue

    timestamp_sec = frame_count / fps
    timestamp_str = f"{int(timestamp_sec//60):02d}:{int(timestamp_sec%60):02d}"

    # Run both models on the frame
    results1 = model1.predict(source=[frame], save=False, verbose=False)
    results2 = model2.predict(source=[frame], save=False, verbose=False)

    # Collect all (results, model) pairs for proper label mapping
    all_results = [(results1, model1), (results2, model2)]

    for results, model in all_results:
        for r in results:
            boxes = getattr(r, "boxes", None)
            if boxes is not None and len(boxes) > 0:
                for box in boxes:
                    class_id = int(box.cls[0])
                    label = model.names[class_id] if hasattr(model, "names") else str(class_id)
                    conf = float(box.conf[0])
                    print(f"[DETECTION @ {timestamp_str}] {label}: {conf:.2f}")

    frame_count += 1

cap.release()
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import time

url = "https://www.ubisoft.com/en-gb/game/rainbow-six/siege/game-info/operators"
operator_urls = [url + "/" + name.lower()  for name in filtered_operator_names]


data_rows = []
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

print(f"Starting to scrape {len(operator_urls)} URLs...")

# --- Main Scraping Loop ---
for url in operator_urls:
    operator_name_slug = url.split('/')[-1].upper()
    print(f"Scraping: {operator_name_slug}...")
    
    # Initialize variables for the current operator
    name, side, squad, specialty = "N/A", "N/A", "N/A", "N/A"
    health, speed, difficulty = 0, 0, 0
    primary_weapon, secondary_weapon, gadget, unique_ability = "N/A", "N/A", "N/A", "N/A"

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status() # Check for bad HTTP response (4xx or 5xx)
        soup = BeautifulSoup(response.text, 'html.parser')

        # --- Extract Header/Name ---
        name_element = soup.find('div', class_="operator__header__icons__names")
        if name_element:
            name = name_element.find("h1").get_text(strip=True)
        
        # --- Extract Side, Squad, Specialty, and Stats ---
        header_info = soup.find('div', class_="operator__header__infos")
        if header_info:
            # Side
            side_div = header_info.find('div', class_='operator__header__side__detail attacker') \
                         or header_info.find('div', class_='operator__header__side__detail defender')
            side = side_div.get_text(strip=True) if side_div else "Unknown"
            
            # Squad and Specialty
            squad = header_info.find('div', class_='operator__header__squad__detail').get_text(strip=True)
            specialty = header_info.find('div', class_='operator__header__roles').get_text(strip=True)

           

            # Health, Speed, Difficulty - FIX APPLIED HERE
            for stat_div in header_info.find_all('div', class_='operator__header__stat'):
                title = stat_div.find('div', class_='operator__header__stat__title').get_text(strip=True).upper()
                
                # CORRECTED STAR COUNT: finding all active stars
                star_count = len(stat_div.find_all('div', class_='react-rater-star is-disabled is-active'))
                
                if "HEALTH" in title:
                    health = star_count
                elif "SPEED" in title:
                    speed = star_count
                elif "DIFFICULTY" in title:
                    difficulty = star_count
        
        # --- Extract Loadout and Ability Data ---
        for item in soup.find_all('div', class_='operator__loadout__category'):
            category_title_element = item.find('h2')
            if not category_title_element:
                continue
                
            category_title = category_title_element.get_text(strip=True).upper()
            
            # Collect all non-empty <p> tags, separating each item with a comma
            data_items = [d.get_text(strip=True) for d in item.find_all('p') if d.get_text(strip=True)]
            data_string = ", ".join(data_items)

            if "PRIMARY WEAPON" in category_title:
                primary_weapon = data_string
            elif "SECONDARY WEAPON" in category_title:
                secondary_weapon = data_string
            elif "GADGET" in category_title:
                gadget = data_string
            elif "UNIQUE ABILITY" in category_title:
                unique_ability = data_string

        
        data_rows.append({
            "Name": name,
            "Side": side,
            "Squad": squad,
            "Specialty": specialty,
            "Health": health,
            "Speed": speed,
            "Difficulty": difficulty,
            "Primary Weapon": primary_weapon,
            "Secondary Weapon": secondary_weapon,
            "Gadget": gadget,
            "Unique Ability": unique_ability
        })

    except requests.exceptions.RequestException as e:
        print(f"Error accessing {url}: {e}")
    except AttributeError as e:
        print(f"Parsing error on {url} (Operator: {operator_name_slug}). Element not found. Error: {e}")
        
    time.sleep(1) # Be polite

# --- Save to CSV ---
csv_filename = "rainbow6_operators_full_data.csv"
if data_rows:
    # Define fieldnames to include all new columns
    fieldnames = [
        "Name", "Side", "Squad", "Specialty", "Health", "Speed", "Difficulty",
        "Primary Weapon", "Secondary Weapon", "Gadget", "Unique Ability"
    ]
    
    with open(csv_filename, "w", newline='', encoding="utf-8") as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data_rows)
        
    print(f"\n✅ Data successfully saved in {csv_filename} with {len(data_rows)} rows.")
else:
    print("\n❌ No data was collected. Check your URLs and selectors.")
```
